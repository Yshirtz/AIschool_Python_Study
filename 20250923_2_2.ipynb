```python
def add_one(n):
    return n+1

result =[]
for i in range(len(my_list)):
    result.append(add_one(my_list[i]))

print(result)

result =[]
result = list(map(add_one,my_list)) 
#map함수도 그대로 출력시키면 주소값만 내오기 때문에 list를 붙였다.
print(result)
```

# 도전문제

```python
#17번 
import random

def checkwin(my_hand,your_hand):
    print(f"나 : {my_hand} 상대 : {your_hand}")
    if my_hand == your_hand:
        print("비겼습니다.")
        return False
    elif my_hand == "가위" and your_hand == "보" or my_hand == "바위" and your_hand == "가위" or my_hand == "보" and your_hand == "바위":
        print("이겼습니다!")
        return True
    else:
        print("졌습니다..")
        return False

def continueConfirm():
    print("계속 하시겠습니까?")
    while True:
        will_continue = input("계속 하시겠습니까? (Y,N) : ")
        if will_continue == "y" or will_continue =="Y":
            return True
        elif will_continue == "n" or will_continue == "N":
            return False
        else:
            continue
    
sel = ["가위", "바위", "보"]
while True:
    user = input("가위 바위 보 : ")
    com = sel[random.randint(0,2)]
    if checkwin(user,com):
        if continueConfirm():
            continue
        else:
            break
```

```python
# 18번

def rank(score):
    result =[]
    avg =[]
    for i,j in score:
        avg.append((i+j)/2)
    avg_sort =sorted(avg,reverse=True)
    for i in avg:
        result.append(avg_sort.index(i)+1)
    return result

score_list = [[80, 70], [70, 80], [30, 50], [90, 100], [100, 90], [100, 100], [10, 30]]
result = rank(score_list)
print(result)
```
