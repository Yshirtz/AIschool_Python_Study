# 함수

틍정 동작을 수행하는 일정한 코드들의 모임, 지금까지 파이썬의 다양한 내장함수를 사용해왔음


```python
#과제
def num_max(a,b,c):
    if a>b and a>c:
        return a
    elif b > c:
        return b
    else:
        return c

num = num_max(0,-9,-10)
print(num)

def num_min(a):
    min = a[0]
    for i in range(1,len(a)):
        if a[i] < min:
            min = a[i]
    return min


num = num_min([72, -4, 98, -51, 0, 88, -23])
print(num)
```

```python
def add_mul(choice, *args):
    print(choice, args)
    if choice == "add":
        sum = 0
        for i in args:
            sum += i
        return sum
    elif choice == "mul":
        sum = 1
        for i in args:
            sum *= i
        return sum
    else:
        print("연산자 잘못됨")
```
