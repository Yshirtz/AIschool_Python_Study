## 최적의 유성우 관측일

```python
# 도시명을 입력하면 관측가능한 유성우에 대한 정보를 알려주는 함수 작성하기

def predict_best_meteor_shower_viewing(city):

    meteor_shower_string = ''# 안내 메시지 초기화

    if ~(cities['city'].str.contains(city,na=False).any()): # 입력한 도시 정보가 존재하지 않으면 오류 메시지
        meteor_shower_string = f'<{city}>를 찾지 못했습니다... 다시 입력해주세요.'
        return meteor_shower_string
    else:# 도시의 위도 정보를 불러오기
        latitud = cities.loc[(cities['city'] ==city),'latitude'].iloc[0]
        # 해당 도시의 위도에서 관측 가능한 별자리 조회
        constel_list =  constellations.loc[(constellations['latitudestart']>=latitud)&(constellations['latitudeend']<=latitud),'constellation'].tolist()
        if not constel_list: # 볼 수 있는 별자리가 존재하지 않으면 오류 메시지
            meteor_shower_string = f'<{city}>에서 관측 가능한 별자리를 찾지 못했습니다... 다시 입력해주세요.'
            return meteor_shower_string
        else: # 별자리 관측 가능 안내 메시지
            print(f"<{city}>에서 관측 가능한 별자리를 찾았습니다! 목록 정리 중..")
    
    for constellation in constel_list: # 도시별 관찰 가능한 별자리 iterate
        # 별자리와 가장 가까운 유성우 조회
        meteor_shower = meteor_showers.loc[meteor_showers['radiant'] ==constellation,'name'].iloc[0]
        # 유성우 관측이 가능한 시작일과 종료일
        meteor_showers_startdate = meteor_showers.loc[meteor_showers['radiant'] ==constellation,'startdate'].iloc[0]
        meteor_showers_enddate = meteor_showers.loc[meteor_showers['radiant'] ==constellation,'enddate'].iloc[0]
        # 유성우가 보일 때의 달의 위상 조회
        moon_phase_list = moon_phases.loc[(moon_phases['date'] >= meteor_showers_startdate) & (moon_phases['date'] <= meteor_showers_enddate)]
        # 달이 눈에 보이는 첫 날 조회
        best_moon_date = moon_phase_list.loc[moon_phase_list['moon percentage'].idxmin()]['date'].strftime("%Y년 %m월 %d일")
        # 사용자에게 정보전달
        meteor_shower_string +=f"\n{meteor_shower}를 보기 위해선\n{constellation}자리를 향해서\n{best_moon_date}에 보시는 걸 추천드립니다.\n"

    return meteor_shower_string
```
